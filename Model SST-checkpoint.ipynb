**Classifying ENSO Type for observational SST data. Utilized Zhiyuan et al. 2019 & Ham et al. 2019 for defining Index parameters**

*Defining Nino Indexes, UCEI, r, & theta values*

```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading in SST data
sst_data = xr.open_dataset('sst.mnmean.nc')

# Longitude slicing for the Nino3, Nino4, and Nino3.4 regions
nino3_region = sst_data.sel(lat=slice(5, -5), lon=slice(270, 330))  # 150°–90° W
nino4_region = sst_data.sel(lat=slice(5, -5), lon=slice(160, 210))  # 160° E–150° W
nino34_region = sst_data.sel(lat=slice(5, -5), lon=slice(190, 240))  # 170°–120° W

# Calculating the mean SST for Nino3, Nino4, and Nino3.4 regions
nino3_index = nino3_region['sst'].mean(dim=['lat', 'lon'])
nino4_index = nino4_region['sst'].mean(dim=['lat', 'lon'])
nino34_index = nino34_region['sst'].mean(dim=['lat', 'lon'])

# Creating a DataFrame with the extracted indices
data = pd.DataFrame({
    'Date': pd.to_datetime(nino3_index['time'].values),
    'Nino3': nino3_index.values,
    'Nino4': nino4_index.values,
    'Nino3.4': nino34_index.values
})

# Handling missing values using interpolation
data['Nino3'].interpolate(method='linear', inplace=True)
data['Nino4'].interpolate(method='linear', inplace=True)
data['Nino3.4'].interpolate(method='linear', inplace=True)

# Normalizing Nino3, Nino4, and Nino3.4 using the equation X' = (X - mean) / stdv
data['Nino3'] = (data['Nino3'] - data['Nino3'].mean()) / data['Nino3'].std()
data['Nino4'] = (data['Nino4'] - data['Nino4'].mean()) / data['Nino4'].std()
data['Nino3.4'] = (data['Nino3.4'] - data['Nino3.4'].mean()) / data['Nino3.4'].std()

# Function to calculate the UCEI
def calculate_ucei(data):
    real_part = data['Nino3'] + data['Nino4']
    imag_part = data['Nino3'] - data['Nino4']
    r = np.sqrt(real_part**2 + imag_part**2)
    
    # Calculating theta based on conditions 
    theta = np.zeros_like(real_part)
    for i in range(len(theta)):
        if real_part[i] > 0:
            theta[i] = np.arctan((data['Nino3'][i] - data['Nino4'][i]) / (data['Nino3'][i] + data['Nino4'][i]))
        elif real_part[i] < 0:
            theta[i] = np.arctan((data['Nino3'][i] - data['Nino4'][i]) / (data['Nino3'][i] + data['Nino4'][i])) - np.pi
    
    ucei = r * np.exp(1j * theta)
    
    # Adding r and theta to the DataFrame
    data['r'] = r
    data['theta'] = theta
    
    return ucei, r, theta

# Function to classify ENSO events
def classify_enso(data):
    ucei, r, theta = calculate_ucei(data)
    
    # Defining ENSO types based on theta and NINO indices
    conditions = [
        (theta > np.radians(15)) & (theta < np.radians(90)),
        (theta > np.radians(-15)) & (theta < np.radians(15)),
        (theta > np.radians(-90)) & (theta < np.radians(-15)),
        (theta > np.radians(-165)) & (theta < np.radians(-90)),
        (theta > np.radians(-195)) & (theta < np.radians(-165)),
        (theta > np.radians(-270)) & (theta < np.radians(-195))
    ]
    
    choices = [
        'EP El Nino',
        'Mixed El Nino',
        'CP El Nino',
        'EP La Nina',
        'Mixed La Nina',
        'CP La Nina'
    ]
    
    data['ENSO_Type'] = np.select(conditions, choices, default='None')
    
    return data

# Classifying ENSO events
data = classify_enso(data)

# Removing rows with 'None' in 'ENSO_Type'
data = data[data['ENSO_Type'] != 'None']

print(data[['Date', 'ENSO_Type']])
```

```python
import cftime

# Function to calculate the UCEI for model data
def calculate_ucei_model(data):
    real_part = data['Model_Nino3'] + data['Model_Nino4']
    imag_part = data['Model_Nino3'] - data['Model_Nino4']
    r = np.sqrt(real_part**2 + imag_part**2)
    
    # Calculating theta based on conditions 
    theta = np.zeros_like(real_part)
    for i in range(len(theta)):
        if real_part[i] > 0:
            theta[i] = np.arctan((data['Model_Nino3'][i] - data['Model_Nino4'][i]) / (data['Model_Nino3'][i] + data['Model_Nino4'][i]))
        elif real_part[i] < 0:
            theta[i] = np.arctan((data['Model_Nino3'][i] - data['Model_Nino4'][i]) / (data['Model_Nino3'][i] + data['Model_Nino4'][i])) - np.pi
    
    ucei = r * np.exp(1j * theta)
    
    # Adding r and theta to the DataFrame
    data['r'] = r
    data['theta'] = theta
    
    return ucei, r, theta

# Function to classify ENSO events for model data
def classify_enso_model(data):
    ucei, r, theta = calculate_ucei_model(data)
    
    # Defining ENSO types based on theta and NINO indices for model data
    conditions = [
        (theta > np.radians(15)) & (theta < np.radians(90)),
        (theta > np.radians(-15)) & (theta < np.radians(15)),
        (theta > np.radians(-90)) & (theta < np.radians(-15)),
        (theta > np.radians(-165)) & (theta < np.radians(-90)),
        (theta > np.radians(-195)) & (theta < np.radians(-165)),
        (theta > np.radians(-270)) & (theta < np.radians(-195))
    ]
    
    choices = [
        'EP El Nino',
        'Mixed El Nino',
        'CP El Nino',
        'EP La Nina',
        'Mixed La Nina',
        'CP La Nina'
    ]
    
    data['ENSO_Type_Model'] = np.select(conditions, choices, default='None')
    
    return data

# List of model files by decade
model_files_by_decade = {
    '1850s': [
        'b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.SST.185001-185912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1021.002.pop.h.SST.185001-185912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1041.003.pop.h.SST.185001-185912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1061.004.pop.h.SST.185001-185912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1081.005.pop.h.SST.185001-185912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1101.006.pop.h.SST.185001-185912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1121.007.pop.h.SST.185001-185912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1141.008.pop.h.SST.185001-185912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1161.009.pop.h.SST.185001-185912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1181.010.pop.h.SST.185001-185912.nc'
    ],
    '1860s': [
        'b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.SST.186001-186912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1021.002.pop.h.SST.186001-186912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1041.003.pop.h.SST.186001-186912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1061.004.pop.h.SST.186001-186912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1081.005.pop.h.SST.186001-186912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1101.006.pop.h.SST.186001-186912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1121.007.pop.h.SST.186001-186912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1141.008.pop.h.SST.186001-186912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1161.009.pop.h.SST.186001-186912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1181.010.pop.h.SST.186001-186912.nc'
    ],
    '1870s': [
        'b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.SST.187001-187912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1021.002.pop.h.SST.187001-187912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1041.003.pop.h.SST.187001-187912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1061.004.pop.h.SST.187001-187912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1081.005.pop.h.SST.187001-187912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1101.006.pop.h.SST.187001-187912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1121.007.pop.h.SST.187001-187912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1141.008.pop.h.SST.187001-187912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1161.009.pop.h.SST.187001-187912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1181.010.pop.h.SST.187001-187912.nc'
    ],
    '1880s': [
        'b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.SST.188001-188912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1021.002.pop.h.SST.188001-188912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1041.003.pop.h.SST.188001-188912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1061.004.pop.h.SST.188001-188912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1081.005.pop.h.SST.188001-188912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1101.006.pop.h.SST.188001-188912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1121.007.pop.h.SST.188001-188912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1141.008.pop.h.SST.188001-188912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1161.009.pop.h.SST.188001-188912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1181.010.pop.h.SST.188001-188912.nc'
    ],
    '1890s': [
        'b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.SST.189001-189912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1021.002.pop.h.SST.189001-189912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1041.003.pop.h.SST.189001-189912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1061.004.pop.h.SST.189001-189912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1081.005.pop.h.SST.189001-189912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1101.006.pop.h.SST.189001-189912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1121.007.pop.h.SST.189001-189912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1141.008.pop.h.SST.189001-189912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1161.009.pop.h.SST.189001-189912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1181.010.pop.h.SST.189001-189912.nc'
    ],
    '1900s': [
        'b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.SST.190001-190912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1021.002.pop.h.SST.190001-190912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1041.003.pop.h.SST.190001-190912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1061.004.pop.h.SST.190001-190912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1081.005.pop.h.SST.190001-190912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1101.006.pop.h.SST.190001-190912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1121.007.pop.h.SST.190001-190912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1141.008.pop.h.SST.190001-190912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1161.009.pop.h.SST.190001-190912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1181.010.pop.h.SST.190001-190912.nc'
    ],
    '1910s': [
        'b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.SST.191001-191912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1021.002.pop.h.SST.191001-191912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1041.003.pop.h.SST.191001-191912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1061.004.pop.h.SST.191001-191912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1081.005.pop.h.SST.191001-191912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1101.006.pop.h.SST.191001-191912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1121.007.pop.h.SST.191001-191912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1141.008.pop.h.SST.191001-191912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1161.009.pop.h.SST.191001-191912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1181.010.pop.h.SST.191001-191912.nc'
    ],
    '1920s': [
        'b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.SST.192001-192912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1021.002.pop.h.SST.192001-192912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1041.003.pop.h.SST.192001-192912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1061.004.pop.h.SST.192001-192912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1081.005.pop.h.SST.192001-192912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1101.006.pop.h.SST.192001-192912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1121.007.pop.h.SST.192001-192912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1141.008.pop.h.SST.192001-192912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1161.009.pop.h.SST.192001-192912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1181.010.pop.h.SST.192001-192912.nc'
    ],
    '1930s': [
        'b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.SST.193001-193912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1021.002.pop.h.SST.193001-193912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1041.003.pop.h.SST.193001-193912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1061.004.pop.h.SST.193001-193912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1081.005.pop.h.SST.193001-193912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1101.006.pop.h.SST.193001-193912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1121.007.pop.h.SST.193001-193912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1141.008.pop.h.SST.193001-193912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1161.009.pop.h.SST.193001-193912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1181.010.pop.h.SST.193001-193912.nc'
    ],
    '1940s': [
        'b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.SST.194001-194912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1021.002.pop.h.SST.194001-194912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1041.003.pop.h.SST.194001-194912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1061.004.pop.h.SST.194001-194912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1081.005.pop.h.SST.194001-194912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1101.006.pop.h.SST.194001-194912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1121.007.pop.h.SST.194001-194912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1141.008.pop.h.SST.194001-194912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1161.009.pop.h.SST.194001-194912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1181.010.pop.h.SST.194001-194912.nc'
    ],
    '1950s': [
        'b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.SST.195001-195912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1021.002.pop.h.SST.195001-195912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1041.003.pop.h.SST.195001-195912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1061.004.pop.h.SST.195001-195912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1081.005.pop.h.SST.195001-195912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1101.006.pop.h.SST.195001-195912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1121.007.pop.h.SST.195001-195912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1141.008.pop.h.SST.195001-195912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1161.009.pop.h.SST.195001-195912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1181.010.pop.h.SST.195001-195912.nc'
    ],
    '1960s': [
        'b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.SST.196001-196912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1021.002.pop.h.SST.196001-196912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1041.003.pop.h.SST.196001-196912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1061.004.pop.h.SST.196001-196912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1081.005.pop.h.SST.196001-196912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1101.006.pop.h.SST.196001-196912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1121.007.pop.h.SST.196001-196912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1141.008.pop.h.SST.196001-196912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1161.009.pop.h.SST.196001-196912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1181.010.pop.h.SST.196001-196912.nc'
    ],
    '1970s': [
        'b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.SST.197001-197912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1021.002.pop.h.SST.197001-197912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1041.003.pop.h.SST.197001-197912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1061.004.pop.h.SST.197001-197912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1081.005.pop.h.SST.197001-197912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1101.006.pop.h.SST.197001-197912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1121.007.pop.h.SST.197001-197912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1141.008.pop.h.SST.197001-197912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1161.009.pop.h.SST.197001-197912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1181.010.pop.h.SST.197001-197912.nc'
    ],
    '1980s': [
        'b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.SST.198001-198912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1021.002.pop.h.SST.198001-198912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1041.003.pop.h.SST.198001-198912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1061.004.pop.h.SST.198001-198912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1081.005.pop.h.SST.198001-198912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1101.006.pop.h.SST.198001-198912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1121.007.pop.h.SST.198001-198912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1141.008.pop.h.SST.198001-198912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1161.009.pop.h.SST.198001-198912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1181.010.pop.h.SST.198001-198912.nc'
    ],
    '1990s': [
        'b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.SST.199001-199912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1021.002.pop.h.SST.199001-199912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1041.003.pop.h.SST.199001-199912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1061.004.pop.h.SST.199001-199912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1081.005.pop.h.SST.199001-199912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1101.006.pop.h.SST.199001-199912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1121.007.pop.h.SST.199001-199912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1141.008.pop.h.SST.199001-199912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1161.009.pop.h.SST.199001-199912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1181.010.pop.h.SST.199001-199912.nc'
    ],
    '2000s': [
        'b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.SST.200001-200912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1021.002.pop.h.SST.200001-200912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1041.003.pop.h.SST.200001-200912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1061.004.pop.h.SST.200001-200912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1081.005.pop.h.SST.200001-200912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1101.006.pop.h.SST.200001-200912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1121.007.pop.h.SST.200001-200912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1141.008.pop.h.SST.200001-200912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1161.009.pop.h.SST.200001-200912.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1181.010.pop.h.SST.200001-200912.nc'
    ],
    '2010s': [
        'b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.SST.201001-201412.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1021.002.pop.h.SST.201001-201412.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1041.003.pop.h.SST.201001-201412.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1061.004.pop.h.SST.201001-201412.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1081.005.pop.h.SST.201001-201412.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1101.006.pop.h.SST.201001-201412.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1121.007.pop.h.SST.201001-201412.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1141.008.pop.h.SST.201001-201412.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1161.009.pop.h.SST.201001-201412.nc',
        'b.e21.BHISTcmip6.f09_g17.LE2-1181.010.pop.h.SST.201001-201412.nc'
    ],
}


# Initializing a dictionary to store combined data for each decade
combined_data_by_decade = {}

# Processing each decade
for decade, files in model_files_by_decade.items():
    all_nino3_indices = []
    all_nino4_indices = []
    all_nino34_indices = []  # List to store Nino3.4 indices
    
    # Processing each model file in the decade
    for file in files:
        model_sst_data = xr.open_dataset(file, chunks={'time': 10})
        
        # Checking if the dataset has valid latitude and longitude coordinates
        if 'TLAT' in model_sst_data.variables and 'TLONG' in model_sst_data.variables:
            lat_coord = 'TLAT'
            lon_coord = 'TLONG'
            lat_dim = 'nlat'
            lon_dim = 'nlon'
        else:
            raise KeyError("The dataset does not have valid latitude and longitude coordinates.")
            
        # Computing the boolean indexers first using .compute()
        lat_indexer_nino3 = ((model_sst_data[lat_coord] >= -5) & (model_sst_data[lat_coord] <= 5)).compute()
        lon_indexer_nino3 = ((model_sst_data[lon_coord] >= 270) & (model_sst_data[lon_coord] <= 330)).compute()
        
        lat_indexer_nino4 = ((model_sst_data[lat_coord] >= -5) & (model_sst_data[lat_coord] <= 5)).compute()
        lon_indexer_nino4 = ((model_sst_data[lon_coord] >= 160) & (model_sst_data[lon_coord] <= 210)).compute()
        
        lat_indexer_nino34 = ((model_sst_data[lat_coord] >= -5) & (model_sst_data[lat_coord] <= 5)).compute()
        lon_indexer_nino34 = ((model_sst_data[lon_coord] >= 190) & (model_sst_data[lon_coord] <= 240)).compute()
        
        # Longitude slicing for the Nino3, Nino4, and Nino3.4 regions
        model_nino3_region = model_sst_data.where(lat_indexer_nino3 & lon_indexer_nino3, drop=True)
        model_nino4_region = model_sst_data.where(lat_indexer_nino4 & lon_indexer_nino4, drop=True)
        model_nino34_region = model_sst_data.where(lat_indexer_nino34 & lon_indexer_nino34, drop=True)
        
        # Calculating the mean SST for Nino3, Nino4, and Nino3.4 regions
        model_nino3_index = model_nino3_region['SST'].mean(dim=[lat_dim, lon_dim])
        model_nino4_index = model_nino4_region['SST'].mean(dim=[lat_dim, lon_dim])
        model_nino34_index = model_nino34_region['SST'].mean(dim=[lat_dim, lon_dim])

        # Appending the indices to the lists
        all_nino3_indices.append(model_nino3_index.values.flatten())
        all_nino4_indices.append(model_nino4_index.values.flatten())
        all_nino34_indices.append(model_nino34_index.values.flatten())
    
# Combining all indices into a single DataFrame for the current decade
    combined_data = pd.DataFrame({
        'Date': model_nino3_index['time'].values,
        'Model_Nino3': np.mean(all_nino3_indices, axis=0),
        'Model_Nino4': np.mean(all_nino4_indices, axis=0),
        'Model_Nino34': np.mean(all_nino34_indices, axis=0)
    })

# Handling missing values using interpolation
    combined_data['Model_Nino3'].interpolate(method='linear', inplace=True)
    combined_data['Model_Nino4'].interpolate(method='linear', inplace=True)
    combined_data['Model_Nino34'].interpolate(method='linear', inplace=True)

# Normalizing Nino3, Nino4, and Nino3.4 using the equation X' = (X - mean) / stdv
    combined_data['Model_Nino3'] = (combined_data['Model_Nino3'] - combined_data['Model_Nino3'].mean()) / combined_data['Model_Nino3'].std()
    combined_data['Model_Nino4'] = (combined_data['Model_Nino4'] - combined_data['Model_Nino4'].mean()) / combined_data['Model_Nino4'].std()
    combined_data['Model_Nino34'] = (combined_data['Model_Nino34'] - combined_data['Model_Nino34'].mean()) / combined_data['Model_Nino34'].std()

# Classifying ENSO events using the combined data
    combined_data = classify_enso_model(combined_data)

# Storing the combined data for the current decade
    combined_data_by_decade[decade] = combined_data

# Combining data from all decades into a single DataFrame
final_combined_data = pd.concat(combined_data_by_decade.values())

# Displaying the classified ENSO events for all decades
print(final_combined_data[['Date', 'ENSO_Type_Model']])
```

```python
# SST data for observational data
sst_data = xr.open_dataset('sst.mnmean.nc')

# Longitude slicing for the Nino3, Nino4, and Nino3.4 regions for observational data
nino3_region_obs = sst_data.sel(lat=slice(5, -5), lon=slice(270, 330))  # 150°–90° W
nino4_region_obs = sst_data.sel(lat=slice(5, -5), lon=slice(160, 210))  # 160° E–150° W
nino34_region_obs = sst_data.sel(lat=slice(5, -5), lon=slice(190, 240))  # 170°–120° W

# Calculating the mean SST for Nino3, Nino4, and Nino3.4 regions for observational data
nino3_index_obs = nino3_region_obs['sst'].mean(dim=['lat', 'lon'])
nino4_index_obs = nino4_region_obs['sst'].mean(dim=['lat', 'lon'])
nino34_index_obs = nino34_region_obs['sst'].mean(dim=['lat', 'lon'])

# Creating a DataFrame with the extracted indices for observational data
data_obs = pd.DataFrame({
    'Date': pd.to_datetime(nino3_index_obs['time'].values),
    'Nino3': nino3_index_obs.values,
    'Nino4': nino4_index_obs.values,
    'Nino3.4': nino34_index_obs.values
})

# Handling missing values using interpolation for observational data
data_obs['Nino3'].interpolate(method='linear', inplace=True)
data_obs['Nino4'].interpolate(method='linear', inplace=True)
data_obs['Nino3.4'].interpolate(method='linear', inplace=True)

# Normalizing Nino3, Nino4, and Nino3.4 using the equation X' = (X - mean) / stdv for observational data
data_obs['Nino3'] = (data_obs['Nino3'] - data_obs['Nino3'].mean()) / data_obs['Nino3'].std()
data_obs['Nino4'] = (data_obs['Nino4'] - data_obs['Nino4'].mean()) / data_obs['Nino4'].std()
data_obs['Nino3.4'] = (data_obs['Nino3.4'] - data_obs['Nino3.4'].mean()) / data_obs['Nino3.4'].std()

# Function to calculate the UCEI for observational data
def calculate_ucei(data):
    real_part = data['Nino3'] + data['Nino4']
    imag_part = data['Nino3'] - data['Nino4']
    r = np.sqrt(real_part**2 + imag_part**2)
    
    # Calculating theta based on conditions 
    theta = np.zeros_like(real_part)
    for i in range(len(theta)):
        if real_part[i] > 0:
            theta[i] = np.arctan((data['Nino3'][i] - data['Nino4'][i]) / (data['Nino3'][i] + data['Nino4'][i]))
        elif real_part[i] < 0:
            theta[i] = np.arctan((data['Nino3'][i] - data['Nino4'][i]) / (data['Nino3'][i] + data['Nino4'][i])) - np.pi
    
    ucei = r * np.exp(1j * theta)
    
    # Adding r and theta to the DataFrame
    data['r'] = r
    data['theta'] = theta
    
    return ucei, r, theta

# Function to classify ENSO events for observational data
def classify_enso(data):
    ucei, r, theta = calculate_ucei(data)
    
    # Defining ENSO types based on theta and NINO indices for observational data
    conditions = [
        (theta > np.radians(15)) & (theta < np.radians(90)),
        (theta > np.radians(-15)) & (theta < np.radians(15)),
        (theta > np.radians(-90)) & (theta < np.radians(-15)),
        (theta > np.radians(-165)) & (theta < np.radians(-90)),
        (theta > np.radians(-195)) & (theta < np.radians(-165)),
        (theta > np.radians(-270)) & (theta < np.radians(-195))
    ]
    
    choices = [
        'EP El Nino',
        'Mixed El Nino',
        'CP El Nino',
        'EP La Nina',
        'Mixed La Nina',
        'CP La Nina'
    ]
    
    data['ENSO_Type'] = np.select(conditions, choices, default='None')
    
    return data

# Classifying ENSO events for observational data
data_obs = classify_enso(data_obs)

# Removing rows with 'None' in 'ENSO_Type' for observational data
data_obs = data_obs[data_obs['ENSO_Type'] != 'None']

# Splitting the dataset into four quadrants based on time period for observational data
time_splits_obs = pd.qcut(data_obs['Date'], q=4)

# Defining seasons for observational data
seasons = {'DJF': [12, 1, 2], 'MAM': [3, 4, 5], 'JJA': [6, 7, 8], 'SON': [9, 10, 11]}
season_bins = [0, 3, 6, 9, 12]
season_labels = ['DJF', 'MAM', 'JJA', 'SON']

# Assigning seasons to each date for observational data
data_obs['Season'] = pd.cut(data_obs['Date'].dt.month % 12 + 1,
                            bins=season_bins,
                            labels=season_labels,
                            ordered=False)

# Calculating the percentage of each ENSO type for each quadrant and season for observational data
enso_counts_quadrants_seasons_obs = data_obs.groupby([time_splits_obs, 'Season'])['ENSO_Type'].value_counts(normalize=True).unstack().fillna(0) * 100

# Calculating the percentage of each ENSO type for each quadrant and season for model data
time_splits_model = pd.qcut(final_combined_data['Date'], q=4)
enso_counts_quadrants_seasons_model = final_combined_data.groupby([time_splits_model, 'Season'])['ENSO_Type_Model'].value_counts(normalize=True).unstack().fillna(0) * 100

# Plotting the bar graph of ENSO type percentages for each quadrant and season comparing model and observed data
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(20, 20), sharey=True)

for i in range(4):
    for j in range(4):
        ax = axes[i][j]
        
        # Plotting model data in red
        if i*4 + j < len(enso_counts_quadrants_seasons_model):
            enso_counts_quadrants_seasons_model.iloc[i*4 + j].plot(kind='bar', ax=ax, color='red', alpha=0.5, position=1, width=0.4)
        
        # Plotting observed data in green
        if i*4 + j < len(enso_counts_quadrants_seasons_obs):
            enso_counts_quadrants_seasons_obs.iloc[i*4 + j].plot(kind='bar', ax=ax, color='green', alpha=0.5, position=0, width=0.4)
        
        left_year = pd.to_datetime(enso_counts_quadrants_seasons_model.index[i*4 + j][0].left).year
        right_year = pd.to_datetime(enso_counts_quadrants_seasons_model.index[i*4 + j][0].right).year
        ax.set_xlabel('ENSO Type')
        ax.set_ylabel('Percentage')
        ax.set_title(f'Percentage of Each ENSO Type ({left_year} to {right_year}, {enso_counts_quadrants_seasons_model.index[i*4 + j][1]})')
        
        # Adding percentage values above each type for model data
        for p in ax.patches[:len(enso_counts_quadrants_seasons_model.columns)]:
            ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='baseline', color='red')
        
        # Adding percentage values above each type for observed data
        for p in ax.patches[len(enso_counts_quadrants_seasons_model.columns):]:
            ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='baseline', color='green')

# Adding a legend to indicate green is observed and red is model
fig.legend(['Model Data', 'Observed Data'], loc='upper right', bbox_to_anchor=(1.1, 1))

plt.tight_layout()
plt.show()
```
