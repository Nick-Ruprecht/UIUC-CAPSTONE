```python
import numpy as np
import pandas as pd
import xarray as xr
from sklearn.utils import resample
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt

# Load the SST data
sst_data = xr.open_dataset('sst.mnmean.nc')

# Longitude slicing for the Nino3 and Nino4 regions
nino3_region = sst_data.sel(lat=slice(5, -5), lon=slice(210, 270))
nino4_region_1 = sst_data.sel(lat=slice(5, -5), lon=slice(160, 210))
nino4_region_2 = sst_data.sel(lat=slice(5, -5), lon=slice(-180, -150))
nino4_region_2 = nino4_region_2.assign_coords(lon=(nino4_region_2.lon + 360) % 360)  # Adjust for 0-360 longitude
nino4_region = xr.concat([nino4_region_1, nino4_region_2], dim='lon')

# Calculate the mean SST for Nino3 and Nino4 regions
nino3_index = nino3_region['sst'].mean(dim=['lat', 'lon'])
nino4_index = nino4_region['sst'].mean(dim=['lat', 'lon'])

# Create a DataFrame with the extracted indices
data = pd.DataFrame({
    'Date': pd.to_datetime(nino3_index['time'].values),
    'Nino3': nino3_index.values,
    'Nino4': nino4_index.values
})

# Handle missing values using interpolation
data['Nino3'].interpolate(method='linear', inplace=True)
data['Nino4'].interpolate(method='linear', inplace=True)

# Normalize Nino3 and Nino4 using the equation X' = (X - mean) / stdv
data['Nino3'] = (data['Nino3'] - data['Nino3'].mean()) / data['Nino3'].std()
data['Nino4'] = (data['Nino4'] - data['Nino4'].mean()) / data['Nino4'].std()

# Function to calculate the UCEI
def calculate_ucei(data):
    real_part = data['Nino3'] + data['Nino4']
    imag_part = data['Nino3'] - data['Nino4']
    r = np.sqrt(real_part**2 + imag_part**2)
    
    # Calculating theta based on conditions 
    theta = np.zeros_like(real_part)
    for i in range(len(theta)):
        if real_part[i] > 0:
            theta[i] = np.arctan((data['Nino3'][i] - data['Nino4'][i]) / (data['Nino3'][i] + data['Nino4'][i]))
        elif real_part[i] < 0 and imag_part[i] > 0:
            theta[i] = np.arctan2((data['Nino3'][i] - data['Nino4'][i]), (data['Nino3'][i] + data['Nino4'][i])) + np.pi
        elif real_part[i] < 0 and imag_part[i] < 0:
            theta[i] = np.arctan2((data['Nino3'][i] - data['Nino4'][i]), (data['Nino3'][i] + data['Nino4'][i])) - np.pi
    
    ucei = r * np.exp(1j * theta)
    
    # Adding r and theta to the DataFrame
    data['r'] = r
    data['theta'] = theta
    
    return ucei, r, theta

# Function to classify El Niño events
def classify_el_nino(data):
    ucei, r, theta = calculate_ucei(data)
    
    # Defining El Niño types based on theta
    data['El_Niño_Type'] = np.where((theta > np.radians(15)) & (theta < np.radians(90)), 'EP-type',
                                    np.where((theta > np.radians(-90)) & (theta < np.radians(-15)), 'CP-type',
                                             np.where((theta > np.radians(-15)) & (theta < np.radians(15)), 'Mixed-type', 'None')))
    
    return data

# Classify El Niño events
data = classify_el_nino(data)

# Display the classified El Niño events without using confidence intervals
print(data[['Date', 'El_Niño_Type']])
```

```python
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt

# Load the SST data
sst_data = xr.open_dataset('sst.mnmean.nc')

# Longitude slicing for the Nino3, Nino4, and Nino3.4 regions
nino3_region = sst_data.sel(lat=slice(5, -5), lon=slice(270, 330))  # 150°–90° W
nino4_region = sst_data.sel(lat=slice(5, -5), lon=slice(160, 210))  # 160° E–150° W
nino34_region = sst_data.sel(lat=slice(5, -5), lon=slice(190, 240))  # 170°–120° W

# Calculate the mean SST for Nino3, Nino4, and Nino3.4 regions
nino3_index = nino3_region['sst'].mean(dim=['lat', 'lon'])
nino4_index = nino4_region['sst'].mean(dim=['lat', 'lon'])
nino34_index = nino34_region['sst'].mean(dim=['lat', 'lon'])

# Create a DataFrame with the extracted indices
data = pd.DataFrame({
    'Date': pd.to_datetime(nino3_index['time'].values),
    'Nino3': nino3_index.values,
    'Nino4': nino4_index.values,
    'Nino3.4': nino34_index.values
})

# Handle missing values using interpolation
data['Nino3'].interpolate(method='linear', inplace=True)
data['Nino4'].interpolate(method='linear', inplace=True)
data['Nino3.4'].interpolate(method='linear', inplace=True)

# Normalize Nino3, Nino4, and Nino3.4 using the equation X' = (X - mean) / stdv
data['Nino3'] = (data['Nino3'] - data['Nino3'].mean()) / data['Nino3'].std()
data['Nino4'] = (data['Nino4'] - data['Nino4'].mean()) / data['Nino4'].std()
data['Nino3.4'] = (data['Nino3.4'] - data['Nino3.4'].mean()) / data['Nino3.4'].std()

# Function to calculate the UCEI
def calculate_ucei(data):
    real_part = data['Nino3'] + data['Nino4']
    imag_part = data['Nino3'] - data['Nino4']
    r = np.sqrt(real_part**2 + imag_part**2)
    
    # Calculating theta based on conditions 
    theta = np.zeros_like(real_part)
    for i in range(len(theta)):
        if real_part[i] > 0:
            theta[i] = np.arctan((data['Nino3'][i] - data['Nino4'][i]) / (data['Nino3'][i] + data['Nino4'][i]))
        elif real_part[i] < 0:
            theta[i] = np.arctan((data['Nino3'][i] - data['Nino4'][i]) / (data['Nino3'][i] + data['Nino4'][i])) - np.pi
    
    ucei = r * np.exp(1j * theta)
    
    # Adding r and theta to the DataFrame
    data['r'] = r
    data['theta'] = theta
    
    return ucei, r, theta

# Function to classify ENSO events
def classify_enso(data):
    ucei, r, theta = calculate_ucei(data)
    
    # Defining ENSO types based on theta and NINO indices
    conditions = [
        (theta > np.radians(15)) & (theta < np.radians(90)),
        (theta > np.radians(-15)) & (theta < np.radians(15)),
        (theta > np.radians(-90)) & (theta < np.radians(-15)),
        (theta > np.radians(-165)) & (theta < np.radians(-90)),
        (theta > np.radians(-195)) & (theta < np.radians(-165)),
        (theta > np.radians(-270)) & (theta < np.radians(-195))
    ]
    
    choices = [
        'EP El Nino',
        'Mixed El Nino',
        'CP El Nino',
        'EP La Nina',
        'Mixed La Nina',
        'CP La Nina'
    ]
    
    data['ENSO_Type'] = np.select(conditions, choices, default='None')
    
    return data

# Classify ENSO events
data = classify_enso(data)

# Remove rows with 'None' in 'ENSO_Type'
data = data[data['ENSO_Type'] != 'None']

print(data[['Date', 'ENSO_Type']])
```

```python
# Split the dataset into four quadrants based on time period
time_splits = pd.qcut(data['Date'], q=4)

# Define seasons
seasons = {'DJF': [12, 1, 2], 'MAM': [3, 4, 5], 'JJA': [6, 7, 8], 'SON': [9, 10, 11]}
season_bins = [0, 3, 6, 9, 12]
season_labels = ['DJF', 'MAM', 'JJA', 'SON']

# Assign seasons to each date
data['Season'] = pd.cut(data['Date'].dt.month % 12 + 1,
                        bins=season_bins,
                        labels=season_labels,
                        ordered=False)

# Calculate the percentage of each ENSO type for each quadrant and season
enso_counts_quadrants_seasons = data.groupby([time_splits, 'Season'])['ENSO_Type'].value_counts(normalize=True).unstack().fillna(0) * 100

# Plotting the bar graph of ENSO type percentages for each quadrant and season
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(20, 20), sharey=True)
colors = ['red', 'orange', 'yellow', 'blue', 'purple', 'cyan']

for i in range(4):
    for j in range(4):
        ax = axes[i][j]
        enso_counts_quadrants_seasons.iloc[i*4 + j].plot(kind='bar', ax=ax, color=colors)
        ax.set_xlabel('ENSO Type')
        ax.set_ylabel('Percentage')
        ax.set_title(f'Percentage of Each ENSO Type ({enso_counts_quadrants_seasons.index[i*4 + j][0].left.year} to {enso_counts_quadrants_seasons.index[i*4 + j][0].right.year}, {enso_counts_quadrants_seasons.index[i*4 + j][1]})')
        
        # Add percentage values above each type
        for p in ax.patches:
            ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='baseline')

plt.tight_layout()
plt.show()
```
